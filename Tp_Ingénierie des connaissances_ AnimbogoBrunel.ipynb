```python
import networkx as nx
import matplotlib.pyplot as plt

G1 = nx.DiGraph()
G1.add_edges_from([
    ("Oiseau", "Animal"),
    ("Oiseau", "Voler"),
    ("Mammifere", "Animal"),
    ("Mammifere", "Allaite"),
    ("Chauve-souris", "Mammifere"),
    ("Chauve-souris", "Oiseau"),
    ("Chauve-souris", "NePondPas")
])
plt.figure(figsize=(6,4))
nx.draw(G1, with_labels=True, node_color="lightblue")
plt.title("Exercice 1")
plt.show()

###2: Voler , Allaiter 
###3 Conflits :La C-S ne pond pas des oeufs alors que les oiseaux pondent des oeufs en general 
```

```python
G3 = nx.DiGraph()
G3.add_edges_from([
    ("Etudiant", "Personne"),
    ("Etudiant", "EtudieDomaine"),
    ("Enseignant", "Personne"),
    ("Enseignant", "EnseigneDomaine"),
    ("Professeur", "Enseignant"),
    ("Professeur", "Chercheur")
])

plt.figure(figsize=(6,4))
nx.draw(G3, with_labels=True, node_color="orange")
plt.title("Exo 3")
plt.show()
```

```python
G4 = nx.DiGraph()
G4.add_edges_from([
    ("Chien", "Animal"),
    ("Chat", "Animal"),
    ("Animal", "EtreVivant"),
    ("Chien", "Os"),
    ("Chat", "Lait"),
    ("Os", "Nourriture")
])

activation = {"Chien": 1.0}
factor = 0.5
levels = 3

current = ["Chien"]
for _ in range(levels):
    next_nodes = []
    for n in current:
        for succ in G4.successors(n):
            activation[succ] = activation.get(succ, 0) + activation[n] * factor
            next_nodes.append(succ)
    current = next_nodes

plt.figure(figsize=(6,4))
nx.draw(G4, with_labels=True, node_color="pink")
plt.title("Exo 4")
plt.show()
```

```python
G7 = nx.DiGraph()
G7.add_edges_from([
    ("Fivre", "Maladie"),
    ("Toux", "MaladieResp"),
    ("Grippe", "MaladieResp"),
    ("Grippe", "Fivre"),
    ("Grippe", "Toux"),
    ("Pneumonie", "MaladieResp"),
    ("Pneumonie", "Fivre"),
    ("Pneumonie", "DouleurPoitrine")
])

symptomes = {"Fivre", "Toux"}

possibles = []
for maladie in ["Grippe", "Pneumonie"]:
    s = set(G7.successors(maladie))
    if symptomes.issubset(s):
        possibles.append(maladie)

print("Maladies possibles avec fivre + toux :", possibles)
```
